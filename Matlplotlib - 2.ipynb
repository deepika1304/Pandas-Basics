```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
Company = ['Amazon', 'Google', 'Flipkart', 'Netflix', 'PayPal']
Revenue = ['90','123','80','78','66']
profit  = ['45','70','40','20','60']
xpos=np.arange(len(Company))
xpos
plt.xticks(xpos,Company)
plt.xlabel("Company")
plt.ylabel("Revenue in Bn")
plt.title("US Stocks")
plt.bar(xpos-0.2, Revenue, width=0.4,label="Revenue")
plt.bar(xpos+0.2, profit, width=0.4, label="profit")
plt.legend()
```

```python
#horizontal bar
plt.yticks(xpos,Company)
plt.barh(xpos-0.2, Revenue,label="Revenue")
plt.barh(xpos+0.2, profit,  label="profit")
plt.legend()
```
