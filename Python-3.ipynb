```python
#Dataframe Different ways of creating dataframe
#Through csv file
import pandas as pd
df = pd.read_csv("financial_year.csv")
df.head()
```

```python
#Dataframe Through excel file
df = pd.read_excel("Student_Details.xlsx")
df.head()
```

```python
#Dataframe through python dictionary
Student_Details = {
    'Name' : ['Aditya', 'Bavin', 'Catherine', 'Dilip'],
    'Year' : ['2000', '1998', '1999', '2001'],
    'Marks' : ['98', '35', '75', '67']
}
df = pd.DataFrame(Student_Details)
df
```

```python
#Dataframe through python Tuple's
Student_Details = [
    ('Aditya', 'Bavin', 'Catherine', 'Dilip'),
    ('2000', '1998', '1999', '2001'),
    ('98', '35', '75', '67')
]
df = pd.DataFrame(Student_Details, columns=["Student_1", "Student_2", "Student_3", "Student_4"])
df
```
