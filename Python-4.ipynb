```python
#To read excel and modify each field data as per our wish using converters
def Name_converter_function(cell):
    if cell=='ABC':
        return 'Modified-ABC'
    return cell

def Empid_converter_function(cell):
    if cell==1:
        return None
    return cell

def Salary_converter_function(cell):
    if cell==10000:
        return 'No-Amount'
    return cell
    
df=pd.read_excel("Data Bleanding File One.xlsx", nrows=10, converters = {
    'NAME': Name_converter_function,
    'EMPID': Empid_converter_function,
    'SALARY': Salary_converter_function
})
df
```

```python
#Copy to other excel
df=pd.read_excel("Data Bleanding File One.xlsx")
#to set offset in excel
df.to_excel("Data-bleanding-2.xlsx",sheet_name='Emplyee Data',startrow=1,startcol=1,index=False)
```

```python
#To write two different dataframes into one excel with two different sheet's

df_details1=pd.DataFrame({
    'Name':['Adithya', 'Bhavan', 'Ceilion', 'Daniel'],
    'Age':['24', '25','26','20'],
    'Year':['2000','1999','1998','2004'],
    'Group':['A+','O+','B+','B-']
})

df_details2=pd.DataFrame({
    'Name':['Eric', 'Fallalh', 'Georgie', 'Hyfa'],
    'Age':['24', '25','26','10'],
    'Year':['2000','1999','1998','2014'],
    'Group':['B+','B+','A+','AB+']
})

with pd.ExcelWriter('Patient_details.xlsx') as details:
    df_details1.to_excel(details,sheet_name="details-1")
    df_details2.to_excel(details,sheet_name="details-2")
```
