```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x_train = np.array([1.0, 2.0, 3.0])
y_train = np.array([200, 300, 400])
m = len(x_train)
i=0
for i in range(m):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x_train[{i}], y_train[{i}]) = ({x_i}, {y_i})")
    i+=1
```

```python
plt.plot(x_train, y_train)
```

```python
w=200
b=200
print(f"w = {w}")
print(f"b = {b}")
```

```python
def calculate_prediction_output(x,w,b):
    f_wb = np.zeros(m)
    for i in range(m):
        x_i = x_train[i]
        f_wb[i] = w*(x_i) + b
       # i+=1
    return f_wb


F_wb = calculate_prediction_output(x_train,w,b)
plt.title("House rents based on size")
plt.xlabel("Size in 1000 sq feet")
plt.ylabel("Rent in 1000 dollars")
plt.plot(x_train, F_wb, label="Prediction output")
plt.plot(x_train, y_train, label="Actual output")
plt.legend()
```
