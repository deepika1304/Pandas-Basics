```python
import matplotlib.pyplot as plt
from importlib import reload
%matplotlib inline 
plt=reload(plt)
a=[1,2,3,4,5]
b=[24,31,30,96,51]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(a,b, color='blue', linewidth=3, linestyle='dotted')
```

```python
m=[1,2,3,4,5]
n=[20,30,40,50,60]
o=[95,16,17,18,89]
p=[78,28,38,48,58]
plt.xlabel('Student Number')
plt.ylabel('Marks')
plt.plot(m,n, label="Max")
plt.plot(m,o, label="Min")
plt.plot(m,p, label="Avg")
plt.legend(loc = "best")
plt.grid()
```
